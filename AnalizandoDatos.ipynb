```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer datos
datos = '/content/sales_data.csv'
datos = pd.read_csv(datos)
```

#Gráfico de barras horizontales: Ingresos Totales por País

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Agrupar los datos por país y calcular los ingresos totales
ingresos_por_pais = datos.groupby('Country')['Revenue'].sum().sort_values(ascending=False)

# Crear gráfico de barras horizontales
plt.figure(figsize=(12, 8))
barras_pais = plt.barh(ingresos_por_pais.index, ingresos_por_pais.values, color='skyblue')

# Agregar los valores debajo de las barras
for barra in barras_pais:
    plt.text(barra.get_width() + 1000, barra.get_y() + barra.get_height() / 2,
             f'{int(barra.get_width()):,}', va='center', ha='left', fontsize=10)

# Función para formatear los valores del eje X
formatter = FuncFormatter(lambda x, pos: f'{int(x):,}')  # Formato con comas como separadores de miles
plt.gca().xaxis.set_major_formatter(formatter)

# Personalización del gráfico
plt.title('Ingresos Totales por País', fontsize=16)
plt.xlabel('Ingresos', fontsize=14)
plt.ylabel('País', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# Gráfico de líneas: Ventas Anuales Totales

```python
from matplotlib.ticker import FuncFormatter


def format_millions(x, pos):
    return f'{int(x):,}'
# Agrupar ventas anuales totales
ventas_por_año = datos.groupby('Year')['Revenue'].sum()

# Crear gráfico de líneas con formato ajustado
plt.figure(figsize=(10, 6))
plt.plot(ventas_por_año.index, ventas_por_año.values, marker='o', color='green')

# Configurar etiquetas del eje x
plt.xticks(ventas_por_año.index, rotation=0)

# Aplicar el formato al eje y
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_millions))

# Títulos y etiquetas
plt.title('Ventas Anuales Totales', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Ingresos Totales', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()



```

#Gráfico de pastel: Distribución por Género

```python


# Calcular el conteo de los géneros directamente
genero_clientes = datos['Customer_Gender'].value_counts()

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(genero_clientes, labels=genero_clientes.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Distribución de Clientes por Género', fontsize=16)

# Obtener los totales
total_clientes = genero_clientes.sum()
total_hombres = genero_clientes['M']
total_mujeres = genero_clientes['F']



# Agregar el texto con los totales debajo del gráfico
plt.tight_layout()

# Agregar los totales debajo del gráfico
plt.figtext(0.5, -0.1, f'Total de Clientes: {total_clientes}\nHombres: {total_hombres}\nMujeres: {total_mujeres}',
            ha='center', fontsize=12, color='black')

plt.show()


```

#:Grafico de barras Horizontales: Ganancias por subcategoría



```python
import matplotlib.ticker as mtick

# Gráfico 3: Ganancias por subcategoría
ganancias_subcategoria = datos.groupby('Sub_Category')['Profit'].sum().sort_values()

plt.figure(figsize=(10, 6))
ganancias_subcategoria.plot(kind='barh', color='mediumseagreen')

# Formato para mostrar números completos en el eje x
plt.gca().xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.title('Ganancias por subcategoría')
plt.xlabel('Ganancias')
plt.ylabel('Subcategoría')
plt.show()

```
